```python
import requests
from bs4 import BeautifulSoup

def scrape_wikipedia_page(url):
    try:
        # Send a GET request to the Wikipedia page
        res = requests.get(url)
        
        # Check if the request was successful
        res.raise_for_status()  # This will raise an HTTPError if the request was not successful
        
        # Parse the HTML content using BeautifulSoup
        soup = BeautifulSoup(res.text, 'html.parser')
        
        # For now, just returning the soup object for further processing
        return soup
    
    except requests.exceptions.HTTPError as err:
        print(f"HTTP error occurred: {err}")  # Handle HTTP errors
    except Exception as err:
        print(f"An error occurred: {err}")  # Handle other potential errors

# Usage
soup = scrape_wikipedia_page('https://en.wikipedia.org/wiki/Rajneesh')

# Example of selective parsing: printing the title of the page
if soup:
    print(soup.title.get_text())
```

```python
heading = soup.find('h1').text
```

```python
heading
```

```python
# Initialize an empty string to store the text content
corpus = ''

# Iterate over all paragraph elements found in the BeautifulSoup object
for p in soup.find_all('p'):
    # Append the text of each paragraph to the corpus string
    corpus += p.text
    # Add a newline character to separate paragraphs
    corpus += '\n'

# Remove any leading or trailing whitespace characters from the corpus string
corpus = corpus.strip()
```

```python
print(corpus)
```

```python
 # Open a file with the name specified by the variable 'heading' and with a '.txt' extension
# The 'w' mode means the file is opened for writing (and will be created if it doesn't exist)
fd = open(heading + '.txt', 'w', encoding= 'utf-8')

# Write the content of the 'corpus' string to the file
fd.write(corpus)

# Close the file to ensure that all data is properly written and resources are released
fd.close()
```
